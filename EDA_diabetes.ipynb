# Diabetesデータセット
## データセット目的
- 糖尿病患者の年齢、性別、肥満度指数、血圧、血清測定のデータから患者にどういった傾向があるのか、また、糖尿病進行度と各データの傾向が発見できるかもしれない。

```python
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling

warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
```

```python
print(diabetes.DESCR)
```

## データセットの概要
### データ数(患者数)
    - 442
### 変数（10個）
    - 年齢(離散量)
    - 性別(離散量)
    - 肥満度指数(連続量)
    - 平均血圧(連続量)
    - 患者ごとの6つの血清測定
### target
    - 患者ごとの基準値から一年後の病気進行尺度(Disease Progress = DP)


- tc:総コレステロール(連続量)
- ldl:低密度リボタンパク質(連続量)悪玉コレステロール
- hdl:高密度リボタンパク質(連続量)善玉コレステロール
- tch:コレステロール？(連続量)
- ltg:中性脂肪？(連続量)
- glu:血糖値(連続量)

### 仮説
- 血糖値が高いと糖尿病進行度は高い
- コレステロールが高いと糖尿病進行度は高い

```python
df = pd.DataFrame(
    diabetes.data,
    columns=['Age', 
             'Sex', 
             'Bmi', 
             'Abp',
             'tc', 
             'ldl',
             'hdl',
             'tch',
             'ltg',
             'glu'])

df['DP'] = diabetes.target


df.head(442)
```

```python
df.info()
```

- データの欠損なし

## カラムごとの解析

### ヒストグラム

```python
for data in df:
    sns.distplot(df[data], kde=False)
    plt.show()
```

- 変数別にみる意味はあまり無い・・・？
- 平均0に標準化されているので実際の数値のどのあたりに多い・少ないのかわからない

### ペアプロット

```python
sns.pairplot(df)
```

ltgとgluが糖尿病の進行度に少なからず関係している可能性がある。

```python
sns.pairplot(df[['Age', 'DP']])
```

### ヒートマップ

```python
sns.heatmap(df.corr(),
           annot=True,
           fmt='.2f')
```

- tcとldlに大きい相関がある
- hdl(善玉コレステロール)と他の血液測定のデータに負の相関がある。

## 分かったこと・気づいたこと

- 糖尿病進行度に関係がありそうなのはBMIと中性脂肪、血糖値。
- 一年後のデータがないと進行具合の原因がわからない？
